```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/calinaryan/Downloads/NN_MB/CDC/CDC2023.csv", low_memory=False)
```

```python
df = df.drop(columns=df.columns[df.isnull().mean() > 1* 0.4])
```

```python
with pd.option_context('display.max_rows', 5, 'display.max_columns', None): 
    display(df)
```

```python
df_mb = df[["raceeth", "q1", "q2", "q3", "q4", "q6", "q7", "q64", "q65", "q8", "q9", "q10", "q11", "q12", "q17",	"q19", "q26", "q27", "q28",	"q29", "q30", "q31", "q35", "q38", "q39", "q46", "q49", "q50", "q51", "q52", "q53", "q54", "q55",	"q58", "q59", "q60", "q61",	"q74", "q75", "q79", "q80", "q81", "q82", "q84", "q85", "q87", "q92"	]]
```

```python
df_mb = df_mb.dropna(thresh=len(df_mb.columns) // 2 + 1)
```

```python
df_mb.isnull().sum()
```

```python
df_mb[["q19", "q26", "q27", "q28", "q31", "q35"]] = df_mb[["q19", "q26", "q27", "q28", "q31", "q35"]].replace({1:2, 2:1})
```

```python
df_mb.head()
```

```python
df_mb[["q30", "q60"]] = df_mb[["q30", "q60"]].replace({2:3, 3:2})

```

```python
df_mb[["q81", "q82"]] = df_mb[["q81", "q82"]].replace({1:3, 3:2, 2:1})
```

```python
df_mb.head()
```

```python
df_mb[["q85"]] = df_mb[["q85"]].replace({1:7, 2:6, 3:5, 5:3, 6:2, 7:1})
```

```python
df_mb[["q75"]] = df_mb[["q75"]].replace({1:8, 2:7, 3:6, 4:5, 5:4, 6:3, 7:2, 8:1})
```

```python
df_mb = df_mb.fillna(df_mb.mean())
```

```python
df_mb.head()
```

```python
df_mb.isnull().sum()
```

```python
normalized_mb=(df_mb-df_mb.min())/(df_mb.max()-df_mb.min())
```

```python
normalized_mb['row_sum'] = normalized_mb.sum(axis=1)
```

```python
normalized_mb.head()
```

```python
normalized_mb['MB'] = normalized_mb.sum(axis=1) > 38*0.5
```

```python
normalized_mb.head()
```

```python
demographic = normalized_mb.astype(float)
```

```python
demographic.info
```

```python
demographic.drop("row_sum", axis=1, inplace=True)
```
